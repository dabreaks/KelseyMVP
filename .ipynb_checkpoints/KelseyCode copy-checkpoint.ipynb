```python
GBmajor= pd.read_csv('Y5 GB Major Season Ag.csv')
GBminor= pd.read_csv('Y5 GB Minor Season Ag.csv')
GBlivestock= pd.read_csv('Y5 GB Livestock.csv')
GBanimalprod= pd.read_csv('Y5 GB Animal Prod.csv')
GBhhcalc= pd.read_csv('Y5 GB hh calc.csv')

MMmajor= pd.read_csv('Y5 MM Major Season Ag.csv')
MMminor= pd.read_csv('Y5 MM Minor Season Ag.csv')
MMlivestock= pd.read_csv('Y5 MM Livestock.csv')
MManimalprod= pd.read_csv('Y5 MM Animal Prod.csv')
MMhhcalc= pd.read_csv('Y5 MM hh calc.csv')

URmajor= pd.read_csv('Y5 UR Major Season Ag.csv')
URminor= pd.read_csv('Y5 UR Minor Season Ag.csv')
URlivestock= pd.read_csv('Y5 UR Livestock.csv')
URanimalprod= pd.read_csv('Y5 UR Animal Prod.csv')
URhhcalc= pd.read_csv('Y5 UR hh calc.csv')

combinednut= pd.read_csv('finalcombined.csv')
```

```python
GBMajor_slice = GBmajor[['hhid', 'q203_c']]
GBMinor_slice = GBminor[['hhid', 'q204_c']]
GBMajor_slice= GBMajor_slice.rename(columns= {'q203_c':'cropcode'})
GBMinor_slice= GBMinor_slice.rename(columns= {'q204_c':'cropcode'})
GBtotal= GBMajor_slice.append(GBMinor_slice).dropna()

MMMajor_slice = MMmajor[['hhid', 'q203_c']]
MMMinor_slice = MMminor[['hhid', 'q204_c']]
MMMajor_slice= MMMajor_slice.rename(columns= {'q203_c':'cropcode'})
MMMinor_slice= MMMinor_slice.rename(columns= {'q204_c':'cropcode'})
MMtotal= MMMajor_slice.append(GBMinor_slice)

URMajor_slice = URmajor[['hhid', 'q203_c']]
URMinor_slice = URminor[['hhid', 'q204_c']]
URMajor_slice= URMajor_slice.rename(columns= {'q203_c':'cropcode'})
URMinor_slice= URMinor_slice.rename(columns= {'q204_c':'cropcode'})
URtotal= URMajor_slice.append(URMinor_slice)
```

```python
for i in pd.unique(GBtotal['cropcode']):
    hhcrop= float(len(pd.unique(GBtotal.hhid[GBtotal.cropcode == i])))
    hhtotal= float(len(pd.unique(GBhhcalc.hhid)))
    print 'Percent of households in GB with', i, 'is', hhcrop / hhtotal
    
print '\n'

for i in pd.unique(MMtotal['cropcode']):
    hhcrop= float(len(pd.unique(MMtotal.hhid[MMtotal.cropcode == i])))
    hhtotal= float(len(pd.unique(MMhhcalc.hhid)))
    print 'Percent of households in MM with', i, 'is', hhcrop / hhtotal

print '\n'

for i in pd.unique(URtotal['cropcode']):
    hhcrop= float(len(pd.unique(URtotal.hhid[URtotal.cropcode == i])))
    hhtotal= float(len(pd.unique(URhhcalc.hhid)))
    print 'Percent of households in UR with', i, 'is', hhcrop / hhtotal
```

```python
GBMajorland = GBmajor[['q203_c', 'land', 'tot_prod']].dropna()
GBMinorland = GBminor[['q204_c', 'land', 'tot_prod']].dropna()
GBMajorland= GBMajorland.rename(columns= {'q203_c':'cropcode'})
GBMinorland= GBMinorland.rename(columns= {'q204_c':'cropcode'})
GBMajorland['land']= GBMajorland['land'] / 2.47105
GBMinorland['land']= GBMinorland['land'] / 2.47105

MMMajorland = MMmajor[['q203_c', 'land', 'tot_prod']]
MMMinorland = MMminor[['q204_c', 'land', 'tot_prod']]
MMMajorland= MMMajorland.rename(columns= {'q203_c':'cropcode'})
MMMinorland= MMMinorland.rename(columns= {'q204_c':'cropcode'})
MMMajorland['land']= MMMajorland['land'] / 2.47105
MMMinorland['land']= MMMinorland['land'] / 2.47105

URMajorland = URmajor[['q203_c', 'land', 'tot_prod']]
URMinorland = URminor[['q204_c', 'land', 'tot_prod']]
URMajorland= URMajorland.rename(columns= {'q203_c':'cropcode'})
URMinorland= URMinorland.rename(columns= {'q204_c':'cropcode'})
URMajorland['land']= URMajorland['land'] / 2.47105
URMinorland['land']= URMinorland['land'] / 2.47105
```

```python
for i in pd.unique(GBMajorland['cropcode']):
    GBMajortotland= GBMajorland['land'][GBMajorland.cropcode == i].sum()
    GBMajortotprod= GBMajorland['tot_prod'][GBMajorland.cropcode == i].sum()
    print 'The yield for', i, 'in the major season in GB is', GBMajortotprod/GBMajortotland, 'kg/ha'
    print 'The total production for', i, 'in the major season is', GBMajorland['tot_prod'][GBMajorland.cropcode == i].sum(), 'kg in GB'
    print 'In the major season the production info for', i, 'is', GBMajorland['tot_prod'][GBMajorland.cropcode == i].describe()
    print 'The total land for', i, 'in the major season is', GBMajorland['land'][GBMajorland.cropcode == i].sum(), 'ha in GB'
    print 'In the major season the land info for', i, 'is', GBMajorland['land'][GBMajorland.cropcode == i].describe()
    print '\n'
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the major season in GB is \n', GBMajortotprod * nutinfo
    print '\n'
    
for i in pd.unique(GBMinorland['cropcode']):
    GBMinortotland= GBMinorland['land'][GBMinorland.cropcode == i].sum()
    GBMinortotprod= GBMinorland['tot_prod'][GBMinorland.cropcode == i].sum()
    print 'The yield for', i, 'in the minor season in GB is', GBMinortotprod/GBMinortotland, 'kg/ha'
    print 'The total production for', i, 'in the minor season is', GBMinorland['tot_prod'][GBMinorland.cropcode == i].sum(), 'kg in GB'
    print 'In the minor season the production info for', i, 'is', GBMinorland['tot_prod'][GBMinorland.cropcode == i].describe()  
    print 'The total land for', i, 'in the minor season is', GBMinorland['land'][GBMinorland.cropcode == i].sum(), 'ha in GB'
    print 'In the minor season the land info for', i, 'is', GBMinorland['land'][GBMinorland.cropcode == i].describe()
    print '\n'  
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the major season in GB is \n', GBMinortotprod * nutinfo
    print '\n'
    
# count: number of cells with i. mean: mean land per farm growing. 

for i in pd.unique(MMMajorland['cropcode']):
    MMMajortotland= MMMajorland['land'][MMMajorland.cropcode == i].sum()
    MMMajortotprod= MMMajorland['tot_prod'][MMMajorland.cropcode == i].sum()
    print 'The yield for', i, 'in the major season in MM is', MMMajortotprod/MMMajortotland, 'kg/ha'
    print 'The total production for', i, 'in the major season is', MMMajorland['tot_prod'][MMMajorland.cropcode == i].sum(), 'kg in MM'
    print 'In the major season the production info for', i, 'is', MMMajorland['tot_prod'][MMMajorland.cropcode == i].describe()
    print 'The total land for', i, 'in the major season is', MMMajorland['land'][MMMajorland.cropcode == i].sum(), 'ha in MM'
    print 'In the major season the land info for', i, 'is', MMMajorland['land'][MMMajorland.cropcode == i].describe()
    print '\n'
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the major season in MM is \n', MMMajortotprod * nutinfo
    print '\n'
    
for i in pd.unique(MMMinorland['cropcode']):
    MMMinortotland= MMMinorland['land'][MMMinorland.cropcode == i].sum()
    MMMinortotprod= MMMinorland['tot_prod'][MMMinorland.cropcode == i].sum()
    print 'The yield for', i, 'in the minor season in MM is', MMMinortotprod/MMMinortotland, 'kg/ha'
    print 'The total production for', i, 'in the minor season is', MMMinorland['tot_prod'][MMMinorland.cropcode == i].sum(), 'kg in MM'
    print 'In the minor season the production info for', i, 'is', MMMinorland['tot_prod'][MMMinorland.cropcode == i].describe()  
    print 'The total land for', i, 'in the minor season is', MMMinorland['land'][MMMinorland.cropcode == i].sum(), 'ha in MM'
    print 'In the minor season the land info for', i, 'is', MMMinorland['land'][MMMinorland.cropcode == i].describe()
    print '\n'
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the minor season in MM is \n', MMMinortotprod * nutinfo
    print '\n'

for i in pd.unique(URMajorland['cropcode']):
    URMajortotland= URMajorland['land'][URMajorland.cropcode == i].sum()
    URMajortotprod= URMajorland['tot_prod'][URMajorland.cropcode == i].sum()
    print 'The yield for', i, 'in the major season in UR is', URMajortotprod/URMajortotland, 'kg/ha'
    print 'The total production for', i, 'in the major season is', URMajorland['tot_prod'][URMajorland.cropcode == i].sum(), 'kg in UR'
    print 'In the major season the production info for', i, 'is', URMajorland['tot_prod'][URMajorland.cropcode == i].describe()
    print 'The total land for', i, 'in the major season is', URMajorland['land'][URMajorland.cropcode == i].sum(), 'ha in UR'
    print 'In the major season the land info for', i, 'is', URMajorland['land'][URMajorland.cropcode == i].describe()
    print '\n'
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the major season in UR is \n', URMajortotprod * nutinfo
    print '\n'
    
for i in pd.unique(URMinorland['cropcode']):
    URMinortotland= URMinorland['land'][URMinorland.cropcode == i].sum()
    URMinortotprod= URMinorland['tot_prod'][URMinorland.cropcode == i].sum()
    print 'The yield for', i, 'in the minor season in UR is', URMinortotprod/URMinortotland, 'kg/ha'
    print 'The total production for', i, 'in the minor season is', URMinorland['tot_prod'][URMinorland.cropcode == i].sum(), 'kg in UR'
    print 'In the minor season the production info for', i, 'is', URMinorland['tot_prod'][URMinorland.cropcode == i].describe()  
    print 'The total land for', i, 'in the minor season is', URMinorland['land'][URMinorland.cropcode == i].sum(), 'ha in UR'
    print 'In the minor season the land info for', i, 'is', URMinorland['land'][URMinorland.cropcode == i].describe()
    print '\n'
    nutinfo= combinednut.iloc[:,7:25][combinednut.name == i]
    print 'The nutritional yield for', i, 'in the minor season in UR is \n', URMinortotprod * nutinfo
    print '\n'
```
